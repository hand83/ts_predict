```python
import pandas as pd
import matplotlib.pyplot as plt
from nsepy import get_history
from nsepy.derivatives import get_expiry_date
from nsepy import get_index_pe_history
from datetime import date
```

```python
%matplotlib inline
```

```python
sorted(get_expiry_date(2019,12))[-1]
```

```python
last_date = sorted(get_expiry_date(2019,12))[-1]
data_tcs = get_history(symbol="TCS", start=date(2018,10,1), end=last_date, 
                       futures=True, expiry_date=last_date)
data_tcs.index = pd.to_datetime(data_tcs.index.values)
```

```python
data_tcs_spot = get_history(symbol="TCS", start=date(2018,10,1), end=last_date)
data_tcs_spot.index = pd.to_datetime(data_tcs_spot.index.values)
```

```python
date_from = max(min(data_tcs.index), min(data_tcs_spot.index))
```

```python
date_from
```

```python
data_tcs.index[data_tcs.index >= date_from]
```

```python
date_from = max(min(data_tcs.index.values), min(data_tcs_spot.index.values))
fig, ax1 = plt.subplots(1, 1, figsize=(16, 8))
ax1.plot(data_tcs.index[data_tcs.index >= date_from], 
         data_tcs.loc[date_from:, 'Close'], label='futures')
ax1.plot(data_tcs_spot.index[data_tcs_spot.index >= date_from], 
         data_tcs_spot.loc[date_from:, 'Close'], label='spot')
ax1.legend()
fig.show()
```
